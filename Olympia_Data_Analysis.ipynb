```python
#Veri analizi için kütüphaneleri yükleyin.
import pandas as pd # veri seti üzerind işlemelr yapmak
import matplotlib.pyplot as plt # veri görselleştirme ve grafik çizme
import seaborn as sns # veri görselleştirme
```

```python
#Çıktıdaki uyarıları kaptaın
import warnings
warnings.filterwarnings('ignore')
```

```python
veri=pd.read_csv('/content/drive/MyDrive/BTK-3-YZ/data/olimpiyatlar_temizlenmis.csv')
print(veri.head())
```

```python
#Saçılım grafiği
sns.scatterplot(x="boy", y="kilo", data=veri) # scatterplot: saçılım grafiği, x ve  y eksenlerindeki sütunlar ve veri seti
plt.title("Boy ve kilo Dağılım Grafiği")#Grafiğin başlığı
plt.show()
```

```python
sns.scatterplot(x="boy", y="kilo", data=veri) # scatterplot: saçılım grafiği, x ve  y eksenlerindeki sütunlar ve veri seti
plt.title("Boy ve kilo Dağılım Grafiği")#Grafiğin başlığı
sns.set_style('darkgrid')
plt.show()
```

```python
#hue gruplandırma için parametre sınıflar için ayrı renklendirme yapar
sns.scatterplot(x="boy", y="kilo",hue="madalya", data=veri) # scatterplot: saçılım grafiği, x ve  y eksenlerindeki sütunlar ve veri seti
plt.title("MAdalyaya Göre Boy ve kilo Dağılım Grafiği")#Grafiğin başlığı
plt.show()
#veri2=veri[veri[:][veri[madalya] == 'gold']]
```

```python
#Regresyon dağılım grafiği
sns.regplot(x="boy", y="kilo",marker='+', scatter_kws={'alpha':0.1}, line_kws={'color':'red'}, data=veri) # scatterplot: saçılım grafiği, x ve  y eksenlerindeki sütunlar ve veri seti
#marker : veri noktasını göstermek için kullanılan karakter
#scatter_kws : veri noktasının saydamlığı
#line_kws : regresyon çizgisinin rengi

plt.title("Boy ve kilo Dağılım Grafiği")#Grafiğin başlığı
plt.show()
```

```python
sns.scatterplot(x="boy", y="kilo",hue="madalya", data=veri, palette='rocket') # scatterplot: saçılım grafiği, x ve  y eksenlerindeki sütunlar ve veri seti
plt.title("Madalyaya Göre Boy ve kilo Dağılım Grafiği")#Grafiğin başlığı
plt.show()
```

```python
sns.lineplot(x="boy", y="kilo", data=veri, color='red') # lineplot: regresyon grafiği
plt.title("Boy ve kilo Çizgi Grafiği")#Grafiğin başlığı
plt.show()
```

```python
sns.lineplot(x="boy", y="kilo",hue='Gender', data=veri, color='red') # lineplot: regresyon grafiği
plt.title("Boy ve kilo Çizgi Grafiği")#Grafiğin başlığı
plt.show()
```

```python
sns.scatterplot(x='boy',palette='Set1', y='kilo', hue='Gender', data=veri)
sns.set_style('darkgrid')
plt.title('Cinsiyete Göre Boy ve Kilo Bağımlılıkları')
plt.show()
```

```python
sns.displot(veri, x='boy') # displot:histogram grafiği
plt.title("Boy Histogram Dağılımı")#Grafiğin başlığı
plt.ylabel('Frekans')
plt.show()
```

```python
sns.displot(veri, x='kilo') # displot:histogram grafiği
plt.title("Kilo Histogram Dağılımı")#Grafiğin başlığı
plt.ylabel('Frekans')
plt.show()
```

```python
sns.displot(veri, x='kilo', hue='Gender') # displot:histogram grafiği
plt.title("Kilo Histogram Dağılımı")#Grafiğin başlığı
plt.ylabel('Frekans')
plt.show()
```

```python
sns.displot(veri, x='boy', hue='Gender') # displot:histogram grafiği
plt.title("Cinsiyet-Boy Histogram Dağılımı")#Grafiğin başlığı
plt.ylabel('Frekans')
plt.show()
```

```python
sns.displot(veri, x="kilo", y="boy", kind="kde")
plt.xlabel("kilo")
plt.ylabel("boy")
plt.title("Kilo - Boy Histogram")
plt.show()
```

```python
sns.displot(veri, x="kilo", y="boy", hue="Gender", kind="kde")
plt.xlabel("kilo")
plt.ylabel("boy")
plt.title("Kilo - Boy Histogram")
plt.show()
```

```python
sns.displot(veri, x="kilo", y="boy", hue="Gender")
plt.fill(True)
plt.xlabel("kilo")
plt.ylabel("boy")
plt.title("Cinsiyete göre Kilo - Boy Histogram")
plt.show()
```

```python
sns.barplot(x="madalya", y="boy", data=veri)
plt.title("Madalyalara göre boy grafikleri")
plt.show()
```

```python
sns.barplot(x="madalya", y="boy", data=veri, hue='Gender')
plt.title("Madalyalara göre boy grafikleri")
plt.show()
```

```python
sns.catplot(x="madalya", y="yas", hue="Gender", col="sezon",
                 data=veri, kind="bar",height=4, aspect=.7)
plt.show()
```

```python
sns.boxplot(x="sezon", y="boy", data=veri)
plt.show()
```

```python
sns.boxplot(x="sezon", y="boy", hue="Gender",data=veri, palette="Set2")
plt.show()
```

```python
sns.catplot(x="sezon", y="boy",hue="Gender", col="madalya",data=veri, kind="box",height=4, aspect=.7)
plt.show()
```

```python
sns.catplot(x="sezon", y="boy",hue="Gender", col="madalya",data=veri, kind="violin",height=4, aspect=.7)
plt.show()
```

```python
sns.heatmap(veri.corr(), annot=True, linewidths=.5, fmt= '.1f')
plt.show()
```

```python
sns.violinplot(x="sezon", y="boy", data=veri, hue="Gender", split=True)
plt.show()
```

```python
sns.catplot(x="sezon", y="boy",
                hue="Gender", col="madalya",
                data=veri, kind="violin", split=True,
                height=4, aspect=.7)
plt.show()
```

```python
g = sns.JointGrid(data=veri, x="kilo", y="boy")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

```python
sns.pairplot(veri)
plt.show()
```

```python
g = sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```
